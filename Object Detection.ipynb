```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
pipeline = kr.pipeline.Pipeline()
```

```python
image_ocr=[
           kr.tools.read(url) for url in['/content/download.jpg','/content/download (2).jpg','/content/download (1).jpg' ]
]
```

```python
len(image_ocr)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[1])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[2])
```

```python
prediction_groups=pipeline.recognize(image_ocr)
```

```python
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, image_ocr, prediction_groups):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
